# Logistic Regression
Load the `mnist` dataset. Split it into training and test sets. Train and test a logistic regression model using scikit-learn. Check the documentation to identify the most important hyperparameters, attributes, and methods of the model. Use them in practice.

## Importing Dependencies

```python
import pandas as pd
import sklearn.model_selection
import sklearn.linear_model
import sklearn.metrics
```

## Importing Dataset

```python
df = pd.read_csv('../../datasets/mnist.csv')
df = df.set_index('id')
df.head(2)
```

## Spliting Dataset  train and test datasets

```python

x = df.drop(['class'], axis=1)
y =df['class']

x_train, x_test, y_train, y_test, = sklearn.model_selection.train_test_split(x,y)

print('df', df.shape)
print('x_train', x_train.shape)
print('x_test', x_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

## Training model

```python
model = sklearn.linear_model.LogisticRegression(max_iter=3000)
model.fit(x_train, y_train);
```

## Testing the trained model

```python

y_predicted = model.predict(x_test)

accuracy = sklearn.metrics.accuracy_score(y_test, y_predicted)
accuracy
```
